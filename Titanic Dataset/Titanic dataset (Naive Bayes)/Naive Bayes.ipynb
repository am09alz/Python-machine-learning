```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
import seaborn as sb
titanic = sb.load_dataset('titanic')
```

```python
titanic
```

```python
df = titanic.drop(['sibsp','parch','embarked','class','who','adult_male','deck','embark_town','alive','alone','survived'],axis=1)
df
```

```python
df_target = titanic[['survived']]
df_target
```

```python
age_dummies = pd.get_dummies(df.sex)
```

```python
df_x = pd.concat([df,age_dummies],axis=1)
df_x
```

```python
df_x.drop('sex',axis=1,inplace=True)
```

```python
df_x.fillna(df_x['age'].mean(),inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x,df_target,test_size=0.2)
model = GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(x_test)
```

```python
model.predict_proba(x_test[:10])
```
