```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = np.random.rand(10,1)
y = 2*x+3 + 0.1*np.random.rand(10,1)

plt.scatter(x,y)
plt.show()
```

```python
x
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[20]])
```

```python
1.98744548*20 + 3.05384358
```

```python
x_test = np.linspace(-3,3)
y_predict = model.predict(x_test[:,None])

plt.scatter(x,y)
plt.plot(x_test,y_predict,color='red')
plt.legend(['Predicted line','Observed data'])
plt.show
```

# More complex Function

```python
x = np.linspace(-5,5,num=100)[:,None]
y = -0.5 + 2.2*x + 0.3*x**3 + 2*np.random.randn(100,1)

plt.plot(x,y)
plt.show()
```

```python
x_new = np.hstack([x,x**2,x**3,x**4])
x_new.shape
```

```python
x.shape
```

```python
model = LinearRegression()
model.fit(x_new,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_new)

plt.scatter(x,y)
plt.plot(x_new[:,0],y_pred,'r')
plt.legend(['Predicted line','Observed data'])
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
x = np.random.rand(10,1) 
y = 2*x+3 + 0.2*np.random.rand(10,1)
```

```python
plt.scatter(x,y)
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.predict([[20]])
```

```python
x_test = np.linspace(-3,3)
y_test = reg.predict(x_test[:,None])
```

```python
plt.scatter(x,y)
plt.plot(x_test,y_test,color='red')
plt.legend(['predicted line','observed data'])
```
